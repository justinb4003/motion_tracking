```python
import pandas as pd
import numpy as np

lp = pd.read_csv('../ballshot/sphere_shot_level.csv')
lp.columns

level1 = lp.iloc[:, 0:5].copy()
level2 = lp.iloc[:, 5:10].copy()
level3 = lp.iloc[:, 10:15].copy()
level4 = lp.iloc[:, 15:20].copy()

runs = [level1, level2, level3, level4]
for df, r in zip(runs, range(len(runs))):
    df.columns = ['time', 'xpos', 'ypos', 'xvel', 'yvel']
    df['run'] = f'level{r+1}'

alldata = pd.concat(runs)
alldata.to_csv('level_runs.csv')
```

```python

for run, r in zip(runs, range(len(runs))):
    run.plot.scatter(x='time', y='xpos', legend=False,
                     title=f'Level shot X position #{r+1}',
                     ylabel='X Position (cm)', xlabel='Time (seconds)').get_figure().savefig(f'level{r+1}_xpos.png')
    run.plot.scatter(x='time', y='ypos', legend=False,
                     title=f'Level shot Y position #{r+1}',
                     ylabel='Y Position (cm)', xlabel='Time (seconds)').get_figure().savefig(f'level{r+1}_ypos.png')
    run.plot.scatter(x='time', y='xvel', legend=False,
                     title=f'Level shot X velocity #{r+1}',
                     ylabel='X Velocity (cm/s)', xlabel='Time (seconds)').get_figure().savefig(f'level{r+1}_xvel.png')
    run.plot.scatter(x='time', y='yvel', legend=False,
                     title=f'Level shot Y velocity #{r+1}',
                     ylabel='Y Velocity (cms)', xlabel='Time (seconds)').get_figure().savefig(f'level{r+1}_yvel.png')
```
